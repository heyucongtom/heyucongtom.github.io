```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
def solve(path): 
    df = pd.read_csv(path)

    plt.plot(df['wavelength'], df['absorbance'], label="Absorbance")
    plt.title("Absorbance Curve of The Unknown Particle")
    plt.legend()
    plt.show()
```

```python
solve("data.csv")
```
